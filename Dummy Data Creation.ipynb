```python
# Problem 1

import numpy as np

cov = np.array([[1.0, 0.8],[0.8,1.0]])
mean = (-3,0)
np.random.seed (0)
pts = np.random.multivariate_normal (mean, cov, size = 500)

print (pts.shape)
```

```python
# Problem 2

import matplotlib.pyplot as plt
plt.plot(pts[:, 0], pts[:, 1], '.', alpha=1)
plt.axis('equal')
plt.title('Scatter')
plt.grid()
plt.show()
```

```python
#Problem 3

xmin, xmax = pts.min(), pts.max()

plt.figure(figsize=(12, 5))

# Histogram for the first dimension
plt.subplot(1, 2, 1)
plt.hist(pts[:, 0], bins=10, edgecolor='black')
plt.title("Dimension 1")
plt.xlim(xmin, xmax)

# Histogram for the second dimension
plt.subplot(1, 2, 2)
plt.hist(pts[:, 1], bins=10, edgecolor='black')
plt.title("Dimension 2")
plt.xlim(xmin, xmax)

plt.show()
```

```python
# Problem 4

import numpy as np

cov = np.array([[1.0, 0.8],[0.8,1.0]])
mean = (0,-3)
np.random.seed (0)
pts2 = np.random.multivariate_normal (mean, cov, size = 500)
plt.figure(figsize=(8, 6))

# Scatter Plot for dataset 0
plt.scatter(pts[:, 0], pts[:, 1], alpha=1, label='0', color='blue')

# Scatter Plot for dataset 1
plt.scatter(pts2[:, 0], pts2[:, 1], alpha=1, label='1', color='orange')

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot")
plt.legend(title='Dataset')
plt.grid(True)
plt.axis("equal")
plt.show()
```

```python
# Problem 5

combined_data = np.vstack((pts, pts2))
print(combined_data.shape)
```

```python
# Problem 6

labels_pts = np.zeros((pts.shape[0], 1))
labels_pts2 = np.ones((pts2.shape[0], 1))

pts_labeled = np.hstack((pts, labels_pts))
pts2_labeled = np.hstack((pts2, labels_pts2))

combined_data = np.vstack((pts_labeled, pts2_labeled))

print("Combined data shape:", combined_data.shape)
print (combined_data)
```
